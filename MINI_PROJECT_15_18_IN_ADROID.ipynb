```python
print(' Mini Project-15: "Convert 100 images to Grey Scale (b/w)" using OpenCV ')
print()


!pip install opencv-python

import cv2
import os
import glob
from google.colab import files
import shutil

# Upload images
uploaded = files.upload()

# Save the uploaded files
for filename in uploaded.keys():
    with open(filename, 'wb') as f:
        f.write(uploaded[filename])

# Create a directory to save the grayscale images
os.makedirs('grayscale_images', exist_ok=True)

# Get the list of uploaded images
image_files = list(uploaded.keys())

for image_file in image_files:
    # Read the image
    image = cv2.imread(image_file)

    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Construct the output filename
    gray_image_filename = os.path.join('grayscale_images', os.path.basename(image_file))

    # Save the grayscale image
    cv2.imwrite(gray_image_filename, gray_image)

    print(f"Converted {image_file} to grayscale and saved as {gray_image_filename}")

# Zip and download the grayscale images
shutil.make_archive('grayscale_images', 'zip', 'grayscale_images')
files.download('grayscale_images.zip')
```

```python
print(' Mini Project-16: "Resize the 100 images to 50% (smaller) " using OpenCV ')
print()

import cv2
import os
from google.colab import files
import shutil

# Upload images
uploaded = files.upload()

# Save the uploaded files
for filename in uploaded.keys():
    with open(filename, 'wb') as f:
        f.write(uploaded[filename])

# Create a directory to save the resized images
os.makedirs('resized_images', exist_ok=True)

# Get the list of uploaded images
image_files = list(uploaded.keys())

for image_file in image_files:
    # Read the image
    image = cv2.imread(image_file)

    # Get the dimensions of the image
    height, width = image.shape[:2]

    # Calculate the new dimensions (50% of original size)
    new_dimensions = (width // 2, height // 2)

    # Resize the image
    resized_image = cv2.resize(image, new_dimensions, interpolation=cv2.INTER_AREA)

    # Construct the output filename
    resized_image_filename = os.path.join('resized_images', os.path.basename(image_file))

    # Save the resized image
    cv2.imwrite(resized_image_filename, resized_image)

    print(f"Resized {image_file} to 50% and saved as {resized_image_filename}")

# Zip and download the resized images
shutil.make_archive('resized_images', 'zip', 'resized_images')
files.download('resized_images.zip')
```

```python
print(' Mini Project-17: "Convert Colored Video to Grey Scale (b/w)" using OpenCV ')
print()

import cv2
from google.colab import files
import os

# Upload video
uploaded = files.upload()

# Save the uploaded video file
for filename in uploaded.keys():
    with open(filename, 'wb') as f:
        f.write(uploaded[filename])
video_filename = list(uploaded.keys())[0]

# Open the video file
cap = cv2.VideoCapture(video_filename)

# Get the width, height, and frames per second (fps) of the video
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('grayscale_video.avi', fourcc, fps, (width, height), False)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Convert the frame to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Write the grayscale frame to the output video
    out.write(gray_frame)

    # Optional: Display the frame (comment out if running on a headless server)
    # cv2.imshow('frame', gray_frame)
    # if cv2.waitKey(1) & 0xFF == ord('q'):
    #     break

# Release everything if job is finished
cap.release()
out.release()
cv2.destroyAllWindows()

# Download the converted video
files.download('grayscale_video.avi')
```
